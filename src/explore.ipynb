````python
import sqlite3


# Crear una conexión a una base de datos SQLite
conn = sqlite3.connect('mydatabase.db')

# Cargar los DataFrames en las tablas SQLite
df_movie.to_sql('movies', conn, if_exists='replace', index=False)
df_credits.to_sql('credits', conn, if_exists='replace', index=False)

# Ejecutar una consulta SQL para unir las tablas y cargar el resultado en un nuevo DataFrame
query = """
SELECT *
FROM movies
JOIN credits ON movies.title = credits.title;
"""
df_resultado = pd.read_sql(query, conn)

# Mostrar el DataFrame resultante con la información unificada
print(df_resultado)

# Cerrar la conexión a la base de datos
conn.close()


df_resultado = df_resultado.loc[:, ~df_resultado.columns.duplicated()]
df_resultado.head()
````

#### Paso 3: Transforma los datos

Como puedes ver, hay algunas columnas con formato JSON. De cada uno de los JSONs, selecciona el atributo `name` y reemplaza las columnas `genres` y `keywords`. Para la columna `cast`, selecciona los tres primeros nombres.

Las únicas columnas que quedan por modificar son `crew` (equipo) y `overview` (resumen). Para la primera columna, transfórmala para que contenga el nombre del director. Para la segunda, conviértela en una lista.

Una vez hayamos terminado de procesar las columnas y que el modelo de recomendación no se confunda, por ejemplo, entre *Jennifer Aniston* y *Jennifer Conelly*, quitaremos los espacios entre las palabras. Aplica esta función a las columnas `genres`, `cast`, `crew` y `keywords`.

Por último, reduciremos nuestro conjunto de datos combinando todas nuestras columnas convertidas anteriores en una sola columna llamada `tags` (que crearemos). Esta columna ahora tendrá todos los elementos separados por comas y luego las reemplazaremos por espacios en blanco. Debería quedar algo así:

````python
#Para extraer el atributo name de los JSONs en las columnas genres, keywords y para seleccionar los tres primeros nombres de la columna cast
# Data transform as expected
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
    
nuevo_df["genres"] = nuevo_df["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
nuevo_df["keywords"] = nuevo_df["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)

nuevo_df["cast"] = nuevo_df["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)

nuevo_df["crew"] = nuevo_df["crew"].apply(lambda x: "".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))

nuevo_df["overview"] = nuevo_df["overview"].apply(lambda x: [x])

nuevo_df.head()
````

````python
import warnings
warnings.filterwarnings('ignore')
````

````python
nuevo_df["overview"] = nuevo_df["overview"].apply(lambda x: [str(x)])
nuevo_df["genres"] = nuevo_df["genres"].apply(lambda x: [str(genre) for genre in x])
nuevo_df["keywords"] = nuevo_df["keywords"].apply(lambda x: [str(keyword) for keyword in x])
nuevo_df["cast"] = nuevo_df["cast"].apply(lambda x: [str(actor) for actor in x])
nuevo_df["crew"] = nuevo_df["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

nuevo_df["tags"] = nuevo_df["overview"] + nuevo_df["genres"] + nuevo_df["keywords"] + nuevo_df["cast"] + nuevo_df["crew"]
nuevo_df["tags"] = nuevo_df["tags"].apply(lambda x: ",".join(x).replace(",", " "))

nuevo_df.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)

nuevo_df.iloc[0].tags
````

#### Paso 4: Construye un KNN

Para resolver este problema crearemos nosotros nuestro propio KNN. Lo primero de todo es vectorizar el texto siguiendo los mismos pasos que aprendiste en la lección anterior.

Una vez lo hayas hecho, tendríamos que elegir una distancia para comparar texto. En este módulo hemos visto algunas, y la única compatible con lo que queremos hacer es la `distancia coseno`:

```py
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

Con este código podremos ver la similitud existente entre nuestros vectores (representaciones vectoriales de la columna `tags`).

Finalmente, podemos diseñar nuestra función de similitud basada en la distancia del coseno. Nuestra propuesta es la siguiente:

```py
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

De tal forma que devolveríamos las 5 películas más similares a la que introduzcamos en el título. Podríamos utilizarla como sigue:

```py
recommend("Introduce una película")
```


````python
# KNN modeling
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(nuevo_df["tags"])

model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(tfidf_matrix)

def get_movie_recommendations(movie_title):
    movie_index = nuevo_df[nuevo_df["title"] == movie_title].index[0]
    distances, indices = model.kneighbors(tfidf_matrix[movie_index])
    similar_movies = [(nuevo_df["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "Batman"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
````
